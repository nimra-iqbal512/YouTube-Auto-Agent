```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
import torch
```

```python
# Use a smaller model for free execution
model_name = "tiiuae/falcon-7b-instruct"

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", torch_dtype=torch.float16)

# Create text generation pipeline
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
# # Define your video title
# title = "How to Start a YouTube Channel in 2025"
prompt = """
Title: How to Start a YouTube Channel in 2025

Write a detailed, engaging YouTube script, of about 1800-2000 words.
Include an introduction, main points, and a conclusion.
Use conversational language and provide examples.

Script:
"""
```

```python
# Generate script
script = generator(
    prompt,
    max_new_tokens=2000,
    num_return_sequences=1,
    do_sample=True,
    temperature=0.9,
    top_p=0.95,
    repetition_penalty=1.2  # Encourages more content
)

script_text = script[0]["generated_text"]
print(script_text)
```

```python
# Save to a file
with open("youtube_script.txt", "w", encoding="utf-8") as file:
    file.write(script_text)

print("Script saved successfully!")
```
