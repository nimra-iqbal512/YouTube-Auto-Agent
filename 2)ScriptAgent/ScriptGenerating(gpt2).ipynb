```python
from transformers import pipeline

def generate_script(title):
    generator = pipeline("text-generation", model="gpt2")
    prompt = f"Title: {title}\n\nScript:"

    script = generator(prompt, max_length=300, num_return_sequences=1)[0]["generated_text"]
    return script

# Example Usage
title = "You have to provide the script for the title 'How to start teaching?'. This script would be for youTube tutorial"
script = generate_script(title)
print(script)
```
